### Step 5 : Animated rendering of the loaded model  <span style="color:red">(mandatory)</span>
* The following cell renders an episodic-animation of the agent (as per the loaded model) on the 16 different test-instances of the environment
* The animation also captures the behaviour of the agent (as per the loaded model)
* The animation for every test-instance is displayed in the form of a 4x4 grid 
* Below the animation , the respective seed values are also mentioned

```python
rows=8
cols=4
grid = GridspecLayout(rows,cols)
index_of_img=0
for i in range(0,rows,2):
    for j in range(cols):
        
        path_of_img=image_paths[index_of_img]
        img=open(path_of_img,'rb').read()
        
        grid[i, j] = widgets.Image(value=img, format='gif', width=220, height=510)
        grid[i+1, j] = widgets.HTML(value=path_of_img[path_of_img.index('IMAGES/')+7 :])
        index_of_img+=1
        

grid
```

### Results and the Analysis (PDF)

When we tested the accuracy of the five different models trained using PG_Agent on 16 different test instances of the environment,we discovered the following results :

* **Model-1** : In some instances the agent consumed a lot of steps in figuring out the center of the landing-pad at coordinates (0,0) even afterlanding

* **Model-2** : And the model encountered the same problem of figuring out the center of the landing-pad.

* **Model-3** : Consumed more number of steps to find suitable trajectory.

* **Model-4** : Model-4 which had lesser neurons per layer (32*16) as compared to previous models was the worst performing model In this model most of the times the agent kept hovering over the landing-pad struggling to figure out the landing-position

* **Model-5** : We tried to increase the learning-rate $\alpha$ to check if we were able to reduce thenumber of steps. It did work out with an $\alpha$ of 0.002. But, success-ratio of the model decreased to 12:16. Since now in certain cases the agent could be seen slipping down the landing-pad


| # | success-ratio | mean-points | mean-steps | Remarks |
| --- | --- | --- | --- | --- |
| Model-1 | 15:16 | 179 | 672 | Good |
| Model-2 | 13:16 | 210 | 547 | Good |
| Model-3 | 16:16 | 164 | 770 | Perfect |
| Model-4 | 4:16 | 48 | 946 | Not Good  |
| Model-5 | 12:16 | 245 | 302 | Good |

* 4 out of 5 models gave good performance and consumed more number of mean-steps **not** replicating the ***best scenario***
